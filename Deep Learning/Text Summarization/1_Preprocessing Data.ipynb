```python
import pandas as pd
import re
```

# Canonical

## Test Data

### Load Data

```python
df_canonical_test = pd.read_csv('./Dataset/df_canonical_test.csv')

df_canonical_test.head()
```

### Remove sentences

```python
df_canonical_test['remove_clean_article'] = df_canonical_test['clean_article'].apply(lambda x: re.sub(r'Liputan6\. com, [^:]+ :', '', x))

df_canonical_test
```

### Save to CSV

```python
df_canonical_test.to_csv('./Dataset/df_canonical_test.csv', index=False)
```
