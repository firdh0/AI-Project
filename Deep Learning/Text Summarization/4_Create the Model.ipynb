# Preprocess Data to Model Inputs

```python
encoder_max_length=512
decoder_max_length=128

def process_data_to_model_inputs(batch):
  # tokenize the inputs and labels
  inputs = tokenizer(batch["remove_clean_article"], padding="max_length", truncation=True, max_length=encoder_max_length)
  outputs = tokenizer(batch["clean_summary"], padding="max_length", truncation=True, max_length=decoder_max_length)

  batch["input_ids"] = inputs.input_ids
  batch["attention_mask"] = inputs.attention_mask
  batch["decoder_input_ids"] = outputs.input_ids
  batch["decoder_attention_mask"] = outputs.attention_mask
  batch["labels"] = outputs.input_ids.copy()

  # because BERT automatically shifts the labels, the labels correspond exactly to `decoder_input_ids`.
  # We have to make sure that the PAD token is ignored
  batch["labels"] = [[-100 if token == tokenizer.pad_token_id else token for token in labels] for labels in batch["labels"]]

  return batch
```

## Preprocess Train Data

```python
tokenized_canonical_train = df_canonical_train_sample.map(
    process_data_to_model_inputs,
    batched=True
)
```

## Combine Encoder and Decoder

```python
# model = EncoderDecoderModel(encoder=encoder, decoder=decoder)

model = EncoderDecoderModel.from_encoder_decoder_pretrained("cahya/bert-base-indonesian-522M", "cahya/bert-base-indonesian-522M")
```

```python
# https://huggingface.co/docs/transformers/main_classes/text_generation

model.config.decoder_start_token_id = tokenizer.cls_token_id
model.config.eos_token_id = tokenizer.sep_token_id
model.config.pad_token_id = tokenizer.pad_token_id
model.config.vocab_size = len(tokenizer.get_vocab())
```

```python
model.config.encoder.vocab_size, len(tokenizer.get_vocab())
```

```python
model.config.max_length = 142
model.config.min_length = 56
model.config.no_repeat_ngram_size = 3
model.config.early_stopping = True
model.config.length_penalty = 2.0
model.config.num_beams = 4
gradient_checkpointing = True
```

```python
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir='/content/drive/MyDrive/Colab Notebooks/Model/',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=2,
    weight_decay=0.01, # regularisasi
    save_total_limit=2,
    num_train_epochs=2,
    fp16=True,
    gradient_accumulation_steps=40,
    eval_steps=2,
)
```

```python
!pip install rouge_score
import datasets

rouge = datasets.load_metric("rouge")

def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.pad_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    rouge_output = rouge.compute(predictions=pred_str, references=label_str, rouge_types=["rouge2"])["rouge2"].mid

    return {
        "rouge2_precision": round(rouge_output.precision, 4),
        "rouge2_recall": round(rouge_output.recall, 4),
        "rouge2_fmeasure": round(rouge_output.fmeasure, 4),
    }
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    tokenizer=tokenizer,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=tokenized_canonical_train['train'],
    eval_dataset=tokenized_canonical_test['test'],
    data_collator=data_collator,
)
```

```python
trainer.train()
```
