# Unigrams, Bigrams, Tigrams

```python
from sklearn.feature_extraction.text import CountVectorizer

import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns
plt.style.use('seaborn')
```

```python
def get_top_ngrams(corpus, ngram_range, stop_words=None, n=None):
    vec = CountVectorizer(stop_words=stop_words, ngram_range=ngram_range).fit(corpus)
    bag_of_words = vec.transform(corpus)

    sum_words = bag_of_words.sum(axis=0)

    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

    common_words = words_freq[:n]
    words = []
    freqs = []
    for word, freq in common_words:
        words.append(word)
        freqs.append(freq)

    df = pd.DataFrame({'Word': words, 'Frequency': freqs})
    return df

def plot_ngrams(df, title):

    plt.figure(figsize=(10, 5))

    ax = sns.barplot(data=df, x='Frequency', y='Word', palette='rocket')

    for p in ax.patches:
        ax.annotate(f'{int(p.get_width())}', (p.get_x() + p.get_width(), p.get_y() + 0.5), ha='right', va='center', color='white')

    plt.title(title)

    plt.show();
```

## Canonical Data Train

### Unigram

```python
stop_words = None
n = 20
unigrams_train = get_top_ngrams(df_canonical_train['remove_clean_article'], (1, 1), stop_words, n)

unigrams_train
```

# Number of letters

```python
# apply(lambda x: pd.Series(list(x))) digunakan untuk membagi setiap kata menjadi huruf-huruf individual
# stack().value_counts() digunakan untuk menghitung jumlah setiap huruf.

from collections import Counter

def count_letters(df):
    letters = df.str.lower().str.replace('[^a-z]', '')
    letter_counts = Counter(''.join(letters))

    # Membuat DataFrame dari hasil perhitungan
    result_df = pd.DataFrame(list(letter_counts.items()), columns=['Word', 'Frequency'])
    result_df = result_df.sort_values(by='Frequency', ascending=False).head(20)

    return result_df
```

# Count the number of sentences, number of words, number of letters in each data

```python
def count_metrics(text):
    sentences = text.split('. ')
    words = text.split()
    characters = len(text)
    return {'Number of sentences': len(sentences), 'Number of words': len(words), 'Number of letters': characters}

def number_description(number):
    for p in number.patches:
      number.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10), textcoords='offset points', color='black')

def graphic_plot(shortest_sentence_train, longest_sentence_train,
                 shortest_word_train, longest_word_train,
                 shortest_letter_train, longest_letter_train, data):
    # Plotting with Seaborn
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

    # Seaborn barplot for Number of sentences
    sns.barplot(x=['Shortest', 'Longest'], y=[shortest_sentence_train, longest_sentence_train], ax=axs[0])
    axs[0].set_title('Number of Sentences')
    number_description(axs[0])

    # Seaborn barplot for Number of words
    sns.barplot(x=['Shortest', 'Longest'], y=[shortest_word_train, longest_word_train], ax=axs[1])
    axs[1].set_title('Number of Words')
    number_description(axs[1])

    # Seaborn barplot for Number of letters
    sns.barplot(x=['Shortest', 'Longest'], y=[shortest_letter_train, longest_letter_train], ax=axs[2])
    axs[2].set_title('Number of Letters')
    number_description(axs[2])

    plt.suptitle(f'Shortest and Longest number of Letters, Words and Sentences in Canonical Data {data}', fontsize=16, y=1.02)

    plt.tight_layout()
    plt.show()
```

## Canonical Data Train

```python
df_canonical_train['metrics'] = df_canonical_train['remove_clean_article'].apply(count_metrics)

# Memisahkan hasil perhitungan menjadi kolom terpisah
df_canonical_train[['Number of sentences', 'Number of words', 'Number of letters']] = pd.DataFrame(df_canonical_train['metrics'].tolist(), index=df_canonical_train.index)

# Menghapus kolom metrics karena sudah tidak diperlukan
df_canonical_train = df_canonical_train.drop('metrics', axis=1)

df_canonical_train.head()
```

```python
shortest_sentence_train = df_canonical_train['Number of sentences'].min()
longest_sentence_train = df_canonical_train['Number of sentences'].max()

shortest_word_train = df_canonical_train['Number of words'].min()
longest_word_train = df_canonical_train['Number of words'].max()

shortest_letter_train = df_canonical_train['Number of letters'].min()
longest_letter_train = df_canonical_train['Number of letters'].max()

graphic_plot(shortest_sentence_train, longest_sentence_train,
             shortest_word_train, longest_word_train,
             shortest_letter_train, longest_letter_train, 'Train')
```
